### reduced by key

```
new_rdd = sc.textFile("/FileStore/logs/*6*") \
  .map(lambda line: line.split()) \
  .map(lambda fields: (fields[2],1)) \
  .reduceByKey(lambda v1,v2: v1+v2)
```

```
new_rdd.take(20)
```

### count by key

```
count_rdd = new_rdd.map(lambda fields: (fields[1],fields[0]))
                  
```

```
count_rdd.countByKey().items()
```
